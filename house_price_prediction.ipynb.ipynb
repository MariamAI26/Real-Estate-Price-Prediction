```python
data = pd.read_csv("/kaggle/input/egypt-housing-rent/houses_data_v2.csv")
data.head()
```

```python
 data["Floor"]= pd.to_numeric(data["Floor"],errors="coerce")
 
```

```python
data.head()
```

```python
data["Floor"]= data["Floor"].fillna(0.0)
```

```python
data.head()
```

```python
data= data.drop(["House_Type","Region","City"],axis = 1)
```

```python
x= data.drop(["Price"],axis = 1)
y= data ["Price"]
```

```python
obj_data = x.select_dtypes(include="object").columns

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in obj_data:
    x[i] = le.fit_transform(x[i])
```

```python
x
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=4)
model.fit(x,y)
model.score(x,y)
```

```python
import pickle

with open("House_Price.pkl","wb") as f:
    pickle.dump(model, f)
```
